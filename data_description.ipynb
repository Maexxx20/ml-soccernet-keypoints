```python
df["height_m"] = df["height_m/ft"].astype(str).str.split("/").str[0]
df["height_m"] = pd.to_numeric(df["height_m"], errors="coerce")

df[["height_m/ft", "height_m"]].head(10)
```

```python
df["height_m"].describe()
```

```python
median_height = df["height_m"].median()
std_height = df["height_m"].std()

median_height, std_height
```

```python
import matplotlib.pyplot as plt

plt.hist(df["height_m"].dropna(), bins=20)
plt.title("Verteilung der Körpergrössen der Athleten")
plt.xlabel("Körpergrösse (Meter)")
plt.ylabel("Anzahl Athleten")
plt.show()
```
